## Train Pretrained YOLOv8

### 1. Load the model

You should already have ultralytics installed, if not use: pip install ultralytics beforehand. Then use the module to import YOLO.

Alternatively, there are other models that can be used such as YOLO8s or YOLO10. For those, see the Ultralytics website

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
```

### 2. Train the model

Since we are locally training, we have agreed upon the following metrics of using 3 epochs. The path of the datafile is very important here and also needs to be updated in settings.yaml if you change its name or otherwise.

```python
results = model.train(data="safety-rd-v1/data.yaml", epochs=3, imgsz=640)
```

### 3. Validate and export

This will automatically use the validation set thats defined in data.yaml and validate the trained model along with exporting all the results to a folder under **~/runs/detect/trainX** where X=number of times the model was trained, for me it was 11. Look for the **results.csv** which has the numbers I mentioned at the beginning along with a confusion matrix and other success metrics. 

Further, exporting will default to a .pt file which saves the trained model for easier future predictions. Inserting arguements into the command allows for custom file type choice.

```python
metrics = model.val()

model.export()
```
